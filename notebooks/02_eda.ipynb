##### Flight Price Distribution

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=travel_data,
    x="distance",
    y="price_flight"
)

plt.title("Distance vs Flight Price")
plt.show()
```

##### Flight Type vs Price

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    data=travel_data,
    x="flightType",
    y="price_flight"
)

plt.xticks(rotation=45)
plt.show()
```

##### Monthly Trend Flight

```python
monthly_price_flight = (
    travel_data.groupby("month")["price_flight"]
    .mean()
    .reset_index()
)

monthly_price_flight
```

```python
plt.figure(figsize=(10,5))

sns.lineplot(
    data=monthly_price_flight,
    x="month",
    y="price_flight"
)

plt.title("Average Flight Price by Month")
plt.show()
```

##### Monthly Trend Hotel

```python
monthly_price_hotel = (
    travel_data.groupby("booking_month")["price_hotel"]
    .mean()
    .reset_index()
)

monthly_price_hotel
```

```python
plt.figure(figsize=(10,5))

sns.lineplot(
    data=monthly_price_hotel,
    x="booking_month",
    y="price_hotel"
)

plt.title("Average Hotel Price by Month")
plt.show()
```

## Flight Price Prediction Analysis

#### Key Findings

1. **Flight distance and travel time exhibit a positive correlation with ticket price**, indicating that longer journeys generally incur higher costs.

2. **Flight type has a significant impact on pricing**, with premium classes showing noticeably higher fares compared to economy-class flights.

3. **Travel agencies display varying pricing patterns**, suggesting that agency-specific pricing strategies influence ticket costs.

4. **Flight price distribution is right-skewed**, indicating the presence of a relatively small number of high-priced flights.

5. **Travel frequency is concentrated among a subset of users**, with the majority of travelers making fewer than 100 trips while only a small percentage exceed this threshold.

6. **Seasonal trends are evident**, with average flight prices peaking during May and June and reaching lower levels in January.

#### Business Impact

- Distance and travel time should be included as primary predictors in the regression model.
- Flight type and agency should be encoded as important categorical features.
- Seasonality features (month, quarter, holiday periods) may improve model performance.

## Travel Seasonality and Behavioral Insights

#### Observation

Flight prices reach their highest levels during **May and June**, while **January** exhibits comparatively lower average prices.

#### Possible Business Explanation

Travel demand in the United States typically increases during June because:

- The academic year ends, enabling family vacations.
- Summer weather encourages leisure travel.
- The peak holiday season begins.
- Longer daylight hours support tourism activities.

#### Business Impact

- Incorporating seasonal indicators into the regression model may improve prediction accuracy.
- Recommendation systems can prioritize popular seasonal destinations during high-demand periods.
- Dynamic pricing models can account for seasonal demand fluctuations.
